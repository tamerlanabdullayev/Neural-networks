```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from collections import Counter
```

```python
data = pd.read_csv("fifa_eda.csv")
headers = data.head()
```

```python
# Frequency vs rating

plt.hist(data['Overall'],bins=20)
plt.xlabel('Overall rating')
plt.ylabel('Frequency')
plt.title('Frequency vs rating')
```

```python
# salary vs rating

data["Salary"] = data["Wage"].astype(float)*1000
plt.plot(data["Overall"], data["Salary"], 'o')
```

```python
nationality_dict = dict(Counter(data["Nationality"]).most_common(10))
# plt.figure(figsize=(16, 10))
plt.bar(nationality_dict.keys(), nationality_dict.values())
```

```python
data_cut = data[["Overall", "Age", "Potential", "Salary"]]
sn.pairplot(data_cut)
```
