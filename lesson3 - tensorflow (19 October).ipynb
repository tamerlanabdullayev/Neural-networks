```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import gradient_descent_v2
```

```python
data = np.loadtxt(r'data.csv', delimiter=',')
# data = np.loadtxt(r'USA_Housing.csv', delimiter=None)
```

```python
x = data[:, :3]
y = data[:, :-1]
```

```python
# training data

x_tr = x[:8000]
y_tr = y[:8000]
```

```python
# test data
x_test = x[8000:, :]
y_test = y[8000:, :]
```

```python
MLP_model = Sequential()
batch_size = 30
epochs = 150
inp_shape = x_tr.shape[1]
print(inp_shape)
```

```python
MLP_model.add(Dense(64, activation='relu', input_shape = (inp_shape,)))
MLP_model.add(Dense(32, activation='relu'))
MLP_model.add(Dense(16, activation='relu'))
MLP_model.add(Dense(1))

MLP_model.compile(loss='mae', optimizer='adam', metrics=['mae'])
history = MLP_model.fit(x_tr, y_tr, batch_size=batch_size, epochs=epochs, verbose=1)
```

```python
prediction = MLP_model.predict(x_test)
plt.plot(prediction, y_test, 'o')
```
